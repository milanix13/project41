# Data Cleaning and Organizing

```python
#List all the state names in the df
grouped = df.groupby('state')
for name,group in grouped:
    print(name)
```

```python
# get list of rows that are not states
non_state_ix = df[(df['state'] == 'AS') | (df['state'] == 'DC') |
                  (df['state'] == 'FSM') | (df['state'] == 'GU') |
                  (df['state'] == 'MP') | (df['state'] == 'NYC') |
                  (df['state'] == 'RMI') | (df['state'] == 'VI')].index
```

```python
# remove non states from the data
df = df.drop(non_state_ix)
df.head()
```

```python
df[['state','tot_cases','tot_death']]
```

```python
#create a pivot table
df_table=pd.pivot_table(df, index='state',values=['tot_cases','tot_death'])
print(df_table)
```

```python
df_table.shape
```

```python
#Rename the column Location of dataframe 2 to state to match with the dataframe 1
df2.rename(columns ={'Location':'state'},inplace= True)
df2
```

```python
##List all the state names in the df2
grouped = df2.groupby('state')
for name,group in grouped:
    print(name)
```

```python
# get list of rows that are not states
non_state_ix2 = df2[(df2['state'] == 'DD2') | (df2['state'] == 'FM') |
                  (df2['state'] == 'IH2') | (df2['state'] == 'MH') |
                  (df2['state'] == 'PR') | (df2['state'] == 'VA2')].index
```

```python
# To drop RP, US, LA and BP2 by using nonStateIndices2 and committing the change with inplace to keep changes.
df2.drop(non_state_ix2, inplace=True)
# Display the new states
df2[['state']]
```

```python
df2[['state','Distributed','Distributed_Janssen','Distributed_Moderna','Distributed_Pfizer']]
```

```python
#create a pivot table
df2_table=pd.pivot_table(df2, index='state',values=['Distributed','Distributed_Janssen','Distributed_Moderna','Distributed_Pfizer'])
print(df2_table)
```

```python
df2_table.shape
```

```python
#Joining the DataFrame df_table and df2_table
df3=pd.merge(df_table, df2_table, on="state",how="inner")
df3
```

```python
#To get the info of new dataframe
df3.describe()
```

```python
# call the dtypes attributes on the df3 to get the data types of each columns.
df3.dtypes
```

```python
# To convert the datatype of each column to longlong type which is easy for viewing. 
df3['tot_cases']=df3['tot_cases'].astype('longlong') 
df3['tot_death']=df3['tot_death'].astype('longlong') 
df3['Distributed']=df3['Distributed'].astype('longlong')
df3['Distributed_Janssen']=df3['Distributed_Janssen'].astype('longlong')
df3['Distributed_Moderna']=df3['Distributed_Moderna'].astype('longlong')
df3['Distributed_Pfizer']=df3['Distributed_Pfizer'].astype('longlong')
df3
```
